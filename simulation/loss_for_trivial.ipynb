```python
import pandas as pd

data = pd.read_pickle('input_data.pickle')
data
```

```python
import random as rd

def trivial_model(): return "away win"

def random_prediction():
    return rd.choice(["home win", "draw", "away win"])

sum_trivial = 0
sum_random = 0 

for id, value in data.iterrows():
    result = value.result
    trivial = trivial_model()
    random_ = random_prediction()

    if result == random_:
        if result == "home win":
            sum_random += value.odds_home - 1
        elif result == "away win":
            sum_random += value.odds_away - 1
        elif result == "draw":
            sum_random += value.odds_draw - 1
        else:
            raise ValueError
    else:
        sum_random += -1
    
    if result == trivial:
        if result == "home win":
            sum_trivial += value.odds_home - 1
        elif result == "away win":
            sum_trivial += value.odds_away - 1
        elif result == "draw":
            sum_trivial += value.odds_draw - 1
        else:
            raise ValueError
    else:
        sum_trivial += -1

profit_trivial = sum_trivial / len(data)
profit_random = sum_random / len(data)

print(profit_random)
print(profit_trivial)
    
```
